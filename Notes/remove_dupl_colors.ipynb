# Remove duplicate colors for countries

#### 7.1.2020

```python
import pandas as pd

# insert your path here:
path = ''
```

```python
df = pd.read_csv(path + 'task_1.csv', usecols=['category', 'color'])
```

```python
df
```

#### drop NaN rows

```python
df = df.dropna()
```

```python
df
```

```python
sorted_list = ['red', 'green', 'blue', 'yellow']
```

```python
df['color_cat'] = pd.Categorical(df['color'], sorted_list)
```

```python
df.drop(columns='color', inplace=True)
```

```python
df = df.sort_values(['category', 'color_cat'])
```

```python
df
```

```python
df_temp = df.drop_duplicates(subset='category')
```

```python
df_new = pd.merge(df, df_temp, how='left', on='category', suffixes=('_orig', '_new'))
```

```python
df_new
```

```python
df_new.drop(columns='color_cat_orig', inplace=True)
df_new.rename(columns={'color_cat_new': 'color'}, inplace=True)
```
